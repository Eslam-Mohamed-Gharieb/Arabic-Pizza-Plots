```python
import matplotlib.pyplot as plt
from mplsoccer import PyPizza
import os
from matplotlib import font_manager as fm, rcParams


fpath2 = os.path.join(rcParams["datapath"], "D:\\Data Analysis Visualizations\\خطوط\\dast-nevis.ttf")
font2 = fm.FontProperties(fname=fpath2)

fpath = os.path.join(rcParams["datapath"], "D:\\Data Analysis Visualizations\\خطوط\\Eslam trial.ttf")
font = fm.FontProperties(fname=fpath)
```

```python
params = ["Attacking output", "Receptions in opp. box", "Shot vol.", "Defending quantity"
         , "Defending quality", "Disrupting opp. moves", "Ball recoveries and interceptions", "Aerial duels"
          "Ball retention ability", "Link-up play volume", "Carry and dribble volume", "Progressive passing"]

#Attacking ouput: عدد الأكشنز والكونتريبوشنز الكمي في صناعة الأهداف المتوقعة لصالح الفريق
params = ["""جﺎﺘﻧﻹﺍ
ﻲﻣﻮﺠﻬﻟﺍ""", """مﻼﺘﺳﻹﺍ
ﻢﺼﺨﻟﺍ ءاﺰﺟ ﺔﻘﻄﻨﻣ ﻲﻓ""", """تﺍﺪﻳﺪﺴﺘﻟﺍ دﺪﻋ""", """دﺪﻋ
ﺔﻴﻋﺎﻓﺪﻟﺍ تﻼﺧﺪﺘﻟﺍ""", """ةﺩﻮﺟ
ﺔﻴﻋﺎﻓﺪﻟﺍ تﻼﺧﺪﺘﻟﺍ""", """ﻞﻴﻄﻌﺗ
ﻢﺼﺨﻟﺍ تﺎﻤﺠﻫ""", """ةﺩﺎﻌﺘﺳﺇ
ةﺮﻜﻟﺍ ضﺍﺮﺘﻋﺇﻭ""", """ﺔﻴﺋﺍﻮﻫ تﺎﻋﺍﺮﺻ""", """ةﺭﺪﻗ
ةﺮﻜﻟﺎﺑ ظﺎﻔﺘﺣﻹﺍ""", """تﺍﺮﻳﺮﻤﺗ
ﺐﻌﻠﻟﺍ ﻂﺑﺭ""", """ﻞﻤﺤﻟﺍ
ﻲﻣﺪﻘﺘﻟﺍ رﻭﺮﻤﻟﺍﻭ""","""ﺔﻴﻣﺪﻘﺗ تﺍﺮﻳﺮﻤﺗ"""]

values = [74, 51, 13, 37, 46, 83, 82, 16, 38, 68, 55, 82]



slice_colors = ["#1A78CF"] * 3 + ["#D70232"] * 5 + ["#FF9300"] * 4
text_colors = ["#000000"] * 3 + ["#F2F2F2"] * 5 + ["#000000"] * 4

# instantiate PyPizza class
baker = PyPizza(
    params=params,                  # list of parameters
    background_color="#222222",     # background color
    straight_line_color="#000000",  # color for straight lines
    straight_line_lw=1,             # linewidth for straight lines
    last_circle_color="#000000",    # color for last line
    last_circle_lw=1,               # linewidth of last circle
    other_circle_lw=0,
    inner_circle_size=2 # linewidth for other circles
)

# plot pizza
fig, ax = baker.make_pizza(
    values,                          # list of values
    figsize=(6, 6.5),                # adjust the figsize according to your need
    color_blank_space="same",        # use the same color to fill blank space
    slice_colors=slice_colors,       # color for individual slices
    value_colors=text_colors,        # color for the value-text
    value_bck_colors=slice_colors,   # color for the blank spaces
    blank_alpha=0.4,                 # alpha for blank-space colors
    kwargs_slices=dict(
        edgecolor="#000000", zorder=2, linewidth=1
    ),                               # values to be used when plotting slices
    kwargs_params=dict(
        color="#F2F2F2", fontsize=11,
        fontproperties=font, va="center"
    ),                               # values to be used when adding parameter labels
    kwargs_values=dict(
        color="#F2F2F2", fontsize=11,
        fontproperties=font, zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="cornflowerblue",
            boxstyle="round,pad=0.2", lw=1
        )
    )                                # values to be used when adding parameter-values labels
)

# add title
fig.text(
    0.515, 0.975, f"{player_name}     -     {player_team}\n", size=16,
    ha="center", fontproperties=font, color="#F2F2F2"
)

# add rectangles
fig.patches.extend([
    plt.Rectangle(
        (0.31, 0.9225), 0.025, 0.021, fill=True, color="#222222",
        transform=fig.transFigure, figure=fig
    ),
    plt.Rectangle(
        (0.462, 0.9225), 0.025, 0.021, fill=True, color="#222222",
        transform=fig.transFigure, figure=fig
    ),
    plt.Rectangle(
        (0.632, 0.9225), 0.025, 0.021, fill=True, color="#222222",
        transform=fig.transFigure, figure=fig
    ),
])
  # these values might differ when you are plotting

# add subtitle
fig.text(
    0.515, 0.955,
    "٢٢/٢١ ﻢﺳﻮﻣ  -   يﺮﺒﻜﻟﺍ ﺲﻤﺨﻟﺍ تﺎﻳﺭﻭﺪﻟﺍ ﻲﻓ ﺰﻛﺮﻤﻟﺍ ﻲﺒﻋﻼﺑ ﺔﻧﺭﺎﻘﻣ يﻮﺌﻤﻟﺍ ﺐﻴﺗﺮﺘﻟﺍ",
    size=13,
    ha="center", fontproperties=font, color="#F2F2F2"
)



plt.savefig('دي كاتيلير.png',facecolor=fig.get_facecolor(), dpi=300, bbox_inches='tight')
```
